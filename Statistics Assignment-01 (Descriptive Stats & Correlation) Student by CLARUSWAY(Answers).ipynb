  ### <span style="color:#49519E">3. Calculate the mean, median, and mode.<span>

```python
# YOUR CODE IS HERE
import numpy as np
from scipy import stats

salary = [300, 250, 100, 125, 300, 300, 150]

mean_salary = np.mean(salary)
print("mean:", mean_salary)

median_salary = np.median(salary)
print("median:", median_salary)

mode_salary = stats.mode(salary)
print("mode:", mode_salary)


```

  ### <span style="color:#49519E">4. Calculate the min, max, Q1, Q3, and interquartile range.<span>

```python
# YOUR CODE IS HERE
import numpy as np
from scipy import stats

salary = [100, 150, 150, 200, 300, 200, 250, 150]

print("min:", np.min(salary))

print("max:", np.max(salary))

print("Q1:", (np.percentile(salary, 25)))

print("Q2:", (np.percentile(salary, 50)))

print("Q3:", (np.percentile(salary, 75)))

print("IQR:", (stats.iqr(salary)))
```

  ### <span style="color:#49519E">5. Calculate the variance and standard deviation.<span>

```python
# YOUR CODE IS HERE
import numpy as np
from scipy import stats

salary = [100, 150, 150, 200, 300, 200, 250, 150]

print("Variance: ", (np.var(salary)))

print("Std: ", (np.std(salary)))

```

  ### <span style="color:#49519E">6. Calculate the skewness and kurtosis.<span>

```python
# YOUR CODE IS HERE
import numpy as np
import scipy.stats as stats

salary = [100, 150, 150, 200, 300, 200, 250, 150]

skewness = stats.skew(salary)
kurtosis = stats.kurtosis(salary)

print("Skewness:", skewness)
print("Kurtosis:", kurtosis)
```

### <span style="color:#49519E">11. Calculate the `Pearson's r` correlation.<span>

```python
# YOUR CODE IS HERE
import pandas as pd
x = pd.Series(range(10, 20))
y = pd.Series([2, 1, 4, 5, 8, 12, 18, 25, 96, 48])

correlation = x.corr(y)
print("Pearson'in r korelasyon katsayısı:", correlation)


```

### <span style="color:#49519E">12. OPTIONAL. Calculate the relevant method to calculate `Spearman's rho` correlation.<span>

```python
# YOUR CODE IS HERE
import pandas as pd
x = pd.Series(range(10, 20))
y = pd.Series([2, 1, 4, 5, 8, 12, 18, 25, 96, 48])

spearman_correlation = x.corr(y, method='spearman')
print("Spearman'in rho korelasyon katsayısı:", spearman_correlation)

```
